```python
import pandas as pd
import numpy as np

trials_judgements = "qrels-final-trials.txt"
trials_raw_judgements = "clinical_trials.judgments.2017.csv"

with open(trials_judgements) as f:
    judgements = f.readlines()

judgements = [x.split() for x in judgements]

judgements = pd.DataFrame(judgements)
judgements.columns = ['topic', 'q0', 'trial', 'relevance']
judgements
```

```python
raw_judgements = pd.read_csv(trials_raw_judgements)
raw_judgements
```

```python
rel_judgements_from_raw = []

def matchGeneRel1(judgement):
    if ((judgement['gene1_annotation_desc'] in ['Exact', 'Missing Variant', 'Different Variant']) or (judgement['gene2_annotation_desc'] in ['Exact', 'Missing Variant', 'Different Variant']) or (judgement['gene3_annotation_desc'] in ['Exact', 'Missing Variant', 'Different Variant'])):
        return True

def matchGeneRel2(judgement):
    if ((judgement['gene1_annotation_desc'] in ['Exact']) or (judgement['gene2_annotation_desc'] in ['Exact']) or (judgement['gene3_annotation_desc'] in ['Exact'])):
        return True
    return False

def getRelevance(judgement):
    if (judgement['pm_rel_desc'] in ['Human PM', 'Animal PM']):
        if ((judgement['disease_desc'] in ['Exact', 'More Specific', 'More General']) & (judgement['demographics_desc'] in ['Matches', 'Not Discussed']) & (judgement['other_desc'] in ['Matches', 'Not Discussed'])):
            if (matchGeneRel1(judgement)):
                if ((judgement['disease_desc'] in ['Exact', 'More Specific']) & (matchGeneRel2(judgement))):
                    return [judgement['trec_topic_number'], '0', judgement['trec_doc_id'], '2']
                else:
                    return [judgement['trec_topic_number'], '0', judgement['trec_doc_id'], '1']

    return [judgement['trec_topic_number'], '0', judgement['trec_doc_id'], '0']

for index, judgement in raw_judgements.iterrows():
    rel_judgements_from_raw.append(getRelevance(judgement))
    
rel_judgements_from_raw = pd.DataFrame(rel_judgements_from_raw)
rel_judgements_from_raw.columns = ['topic', 'q0', 'trial', 'relevance']
len(rel_judgements_from_raw)
```
